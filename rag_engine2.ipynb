```python
# file uploader

def file_loader(folder_path:str):
    all_doc = []
    for filename in os.listdir.folder_path:
        if filename.endswith('.pdf'):
            path = os.path.join(folder_path,filename)
            loader = PyPDFLoader()
            docs= loader.load()
            all_doc.extend(docs)

            for doc in docs:
                doc.metadata['filename'] = filename
        else:
            print('upload a valid doc or path')
    return all_doc
```
